## 处理population 2017文件

```python
import pandas as pd
df = pd.read_csv('./geo/population_summary_lga_2017.csv')
df
```

```python

print(df.columns)
```

```python
victoria_data = df[df[' state_name'] == 'Victoria']
victoria_data 
```

```python
# 去掉括号
desired_columns = [' persons_num', ' lga_name17']
filtered_vic_df = victoria_data[desired_columns]
filtered_vic_df[' lga_name17'] = df[' lga_name17'].str.replace(r'\(.*?\)', '', regex=True)

# 去掉多余空白
filtered_vic_df[' lga_name17'] = filtered_vic_df[' lga_name17'].str.strip()

filtered_vic_df.to_csv('./geo/population_2017.csv', index = False)
```

```python
# 小写改大写
df = pd.read_csv('./geo/population_2017.csv')
df[' lga_name17'] = df[' lga_name17'].str.upper()

# 去掉多余空格
df = df.rename(columns={' persons_num': 'persons_num', ' lga_name17': 'lga_name17'})
```

```python
df.to_csv('./geo/population_2017.csv', index = False)
```
