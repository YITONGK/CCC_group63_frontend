```python
import geopandas as gpd
import folium

# 载入 Shapefile
shape_path = './geo/vic_localities.shp'
vic_suburbs = gpd.read_file(shape_path)
vic_state = gpd.read_file('./geo/victoria_state.shp')

print(vic_suburbs.columns)
```

```python
print(vic_suburbs['LOC_NAME'])
```

```python
# 用户输入的郊区名
suburb_name = "Abbeyard"

# 筛选出指定的郊区
selected_suburb = vic_suburbs[vic_suburbs['LOC_NAME'] == suburb_name]

# 计算选定郊区的边界
bounds = selected_suburb.total_bounds
lat_center = (bounds[1] + bounds[3]) / 2
lon_center = (bounds[0] + bounds[2]) / 2

# 创建地图，定位到选定郊区的中心
map_victoria = folium.Map(location=[lat_center, lon_center], zoom_start=12)  # 可以调整zoom_start以获得最佳视图

# 为Victoria添加蓝色边界
folium.GeoJson(
    vic_state,
    name='Victoria Boundary',
    style_function=lambda feature: {
        'fillColor': 'blue',  # 蓝色填充
        'color': 'blue',      # 蓝色边界线
        'weight': 2,          # 边界线宽度
        'fillOpacity': 0.2    # 填充透明度
    },
).add_to(map_victoria)

# 为选定的郊区添加红色边界
folium.GeoJson(
    selected_suburb,
    name='Selected Suburb',
    style_function=lambda feature: {
        'fillColor': '#ffaf00',
        'color': 'red',
        'weight': 5,
        'fillOpacity': 0.5
    },
    tooltip=suburb_name
).add_to(map_victoria)

# 显示地图
map_victoria
```
