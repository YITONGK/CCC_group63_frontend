```python
import pandas as pd
```

```python
df_accident = pd.read_csv('./geo/ACCIDENT.csv')
df_accident = df_accident[df_accident['ACCIDENT_DATE'] == 'Victoria']
```

```python
# 确保ACCIDENT_DATE是日期格式
df_accident['ACCIDENT_DATE'] = pd.to_datetime(df_accident['ACCIDENT_DATE'], format='%Y-%m-%d', errors='coerce')

# 过滤日期在2022年或2023年的数据
filtered_data = df_accident[(df_accident['ACCIDENT_DATE'].dt.year == 2022) | (df_accident['ACCIDENT_DATE'].dt.year == 2023)]
filtered_data.to_csv('./geo/accident_22-23.csv', index=False)
```

```python
# 车祸数据
accidents_data = pd.read_csv('geo/ACCIDENT.csv')
accidents_location_data = pd.read_csv('geo/accident_location.csv')

# 将ACCIDENT_DATE列转换为日期
accidents_data['ACCIDENT_DATE'] = pd.to_datetime(accidents_data['ACCIDENT_DATE'])
# 筛选出2022年和2023年的数据
filtered_data = accidents_data[(accidents_data['ACCIDENT_DATE'].dt.year == 2022) | (accidents_data['ACCIDENT_DATE'].dt.year == 2023)]
# 选择需要的列
filtered_data = filtered_data[['ACCIDENT_NO', 'ACCIDENT_DATE', 'SEVERITY', 'SPEED_ZONE']]
# 使用ACCIDENT_NO进行匹配，将两个数据集合并
merged_data = pd.merge(filtered_data, accidents_location_data, on='ACCIDENT_NO')
# 创建新的location列
merged_data['location'] = list(zip(merged_data['LATITUDE'], merged_data['LONGITUDE']))
merged_data = merged_data.drop(columns=['LATITUDE', 'LONGITUDE'])

print(merged_data.head())
print(merged_data.shape)
print(filtered_data.shape)
```

```python
merged_data.to_csv('./geo/accident_location.csv', index=False)
merged_data
```
